```python
#data visualization Q1
plt.plot(df["Age"],df["Experience_Years"], marker='o', linestyle=':', color='pink')

plt.title("Age VS Experience Years")
plt.xlabel("Age")
plt.ylabel("Experience Years")
plt.grid(True)
plt.show()
```

```python
#data visualization Q2
plt.bar(df["Education_Level"],df["Department"], color='blue', width=0.5)

plt.title("Department VS Education_Level")
plt.xlabel("Department")
plt.ylabel("Education_Level")
plt.show()
```

```python
#data visualization Q3
plt.hist(df["Gender"],bins=2, color='green', edgecolor='black')
plt.title("Gender")
plt.xlabel("Gender")
plt.ylabel("")
plt.locator_params(axis="y", integer=True)#to give me y elements as Integer
plt.show()
```

```python
#data visualization Q4
df["Department"].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Employee by Departments")
plt.ylabel("")  # Hide default y-label
plt.show()
```

```python
plt.scatter(df["Experience_Years"],df["Salary"], edgecolor='black', marker='^', color='gold')

plt.title("Experience Years VS Salary")
plt.xlabel("Employee Experience Years")
plt.ylabel("Employee Salary")
plt.show()
```
