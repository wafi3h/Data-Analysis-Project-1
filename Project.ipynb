```python
plt.plot(df["Age"],df["ExperienceInCurrentDomain"], marker='o', linestyle=':', color='pink')

plt.title("Age VS Experience Years")
plt.xlabel("Age")
plt.ylabel("Experience In Current Domain")
plt.grid(True)
plt.show()
```

```python
plt.bar(df["PaymentTier"],df["ExperienceInCurrentDomain"], color='blue', width=0.5)

plt.title("Payment Tier VS Experience In Current Domain")
plt.xlabel("PaymentTier")
plt.ylabel("ExperienceInCurrentDomain")
plt.show()
```
