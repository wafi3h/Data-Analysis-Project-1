```python
join_counts = df["JoiningYear"].value_counts().sort_index()

plt.plot(join_counts.index, join_counts.values, marker="o", linestyle=":", color="teal")
plt.title("Joining Year vs Number of Employees")
plt.xlabel("Joining Year")
plt.ylabel("Number of Employees")
plt.grid(True)
plt.show()
```

```python
joinYears = df['JoiningYear'].value_counts().sort_index()  # count employees per year

joinYears.plot(kind='bar', color='blue')  # bar chart

plt.title('Number of Employees by Joining Year')
plt.xlabel('Joining Year')
plt.ylabel('Number of Employees')
plt.xticks(rotation=20)
plt.show()
```

```python
topCities = df['City'].value_counts()
topCities.plot(kind='bar')
plt.title('The top 10 cities of employees')
plt.xlabel('City')
plt.ylabel('Number of ')
plt.xticks(rotation=20)
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()  # count number of employees for each gender

gender_counts.plot(kind='bar', color='red', edgecolor='black')
plt.title('Number of Employees by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

```python
df["City"].value_counts().plot.pie(autopct="%1.1f%%", startangle = 180) #start from corner 180 
plt.title("Employee by Location")
plt.ylabel("")  # Hide default y-label
plt.show()   
```

```python
plt.scatter(df["ExperienceInCurrentDomain"], df["PaymentTier"], color="red", edgecolor="black")  # scatter plot

plt.title("Experience vs Payment Tier")
plt.xlabel("Experience in Current Domain (years)")
plt.ylabel("Payment Tier")
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df["JoiningYear"])
plt.title("Box Plot of Joining Year for emploees")
plt.ylabel("Joining Year")
plt.show()
```

```python
corr = df.select_dtypes(include="number").corr()

plt.figure(figsize=(8,6))
plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)  # ألوان من -1 إلى 1
plt.colorbar(label="Correlation")  # شريط الألوان

# تسمية الأعمدة على المحاور
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

# كتابة قيمة الارتباط داخل كل مربع
for i in range(len(corr)):
    for j in range(len(corr)):
        plt.text(j, i, f"{corr.iloc[i,j]:.2f}", ha='center', va='center', color="black", fontsize=12)

plt.title("Correlation Heatmap for Numeric Columns")
plt.tight_layout()
plt.show()
```
